```python
canvas_coord = np.array([[566, 420, 1], [654, 408, 1],[566, 609, 1], [654, 578, 1]])
texture_coord = np.array([[0, 0, 1],[646, 0, 1], [0, 284, 1], [646, 284, 1]])
```

```python
print(canvas_coord)
print(texture_coord)
```

```python
t_helper = np.zeros((8,8))
B_helper = np.zeros((8,1))
```

```python
for i in range(4):
    t_helper[i][0] = texture_coord[i][0]
    t_helper[i][1] = texture_coord[i][1]
    t_helper[i][2] = max(1, i)
    
    t_helper[4+i][3] = texture_coord[i][0]
    t_helper[4+i][4] = texture_coord[i][1]
    t_helper[4+i][5] = max(1, i)
    
    t_helper[i][6] = -1*texture_coord[i][0]*canvas_coord[i][0]
    t_helper[i][7] = -1*texture_coord[i][1]*canvas_coord[i][0]
    
    t_helper[4+i][6] = -1*texture_coord[i][0]*canvas_coord[i][1]
    t_helper[4+i][7] = -1*texture_coord[i][1]*canvas_coord[i][1]

for i in range(4):
    B_helper[i][0] = canvas_coord[i][0]
    B_helper[4+i][0] = canvas_coord[i][1]
```

```python
print(t_helper, B_helper)
```

```python
sol_helper = np.linalg.solve(t_helper, B_helper)
print(sol_helper)

transform = np.zeros((3,3))

for i in range(8):
    transform[i//3][i%3] = sol_helper[i][0]

transform[2][2] = 1

print(transform)
```

```python
aux = np.matmul(transform, texture_coord[3])
aux = aux/aux[2]

print(texture_coord[3])
print(aux)
```

```python
inv_transform = np.linalg.inv(transform)
print(inv_transform)
print(np.matmul(inv_transform, canvas_coord[0])/0.31807)
```
